```python
import pandas as pd

# bond와 primary_drop이라는 두 DataFrame이 미리 정의되어 있고, 두 DataFrame 모두 '종목코드' 열을 포함하고 있다고 가정합니다.

# 먼저 '기준일자' 열을 primary_drop DataFrame에서 제거합니다.
# 이 작업을 반복문 밖에서 한 번만 수행하여 효율성을 높입니다.
# primary_drop = primary_drop.drop('기준일자', axis=1)

# bond DataFrame을 업데이트한 행으로 저장할 새로운 DataFrame을 생성합니다.
updated_bond = pd.DataFrame()

for row in bond.iterrows():
    primary_row = primary_drop[primary_drop['종목코드'] == row[1]['종목코드']]  # 같은 종목코드를 가진 primary_drop의 행들
    # print(primary_row)
    
    merge_row = pd.DataFrame(row[1]).T
    
    updated_rows = pd.merge(merge_row, primary_row, on='종목코드', how='inner')    
     
    # 이 결합된 행들을 새 DataFrame에 추가합니다.
    updated_bond = pd.concat([updated_bond, updated_rows], axis=0)

    # print(updated_bond)

# 최종적으로 updated_bond가 원래의 bond 행들과 primary_drop에서 일치하는 행들을 포함하게 됩니다.
bond = updated_bond
```

```python
def monthToQlMonth(month):
    if month == '01':
        return ql.January
    elif month == '02':
        return ql.February
    elif month == '03':
        return ql.March
    elif month == '04':
        return ql.April
    elif month == '05':
        return ql.May
    elif month == '06':
        return ql.June
    elif month == '07':
        return ql.July
    elif month == '08':
        return ql.August
    elif month == '09':
        return ql.September
    elif month == '10':
        return ql.October
    elif month == '11':
        return ql.November
    elif month == '12':
        return ql.December
    else:
        return None

def strToQlDate(dateStr):
    return ql.Date(int(dateStr[6:]), monthToQlMonth(dateStr[4:6]), int(dateStr[:4]))

# 한국 달력 지정
calendar1 = ql.SouthKorea()

def calculateQlDiffDays(date1, date2):
    return calendar1.businessDaysBetween(date1, date2, includeFirst=True, includeLast=False)
```

```python
except_coco['ql_거래일자'] = except_coco['거래일자'].astype(str).apply(strToQlDate)
except_coco['ql_상환일자'] = except_coco['상환일자'].astype(str).apply(strToQlDate)

except_coco
```

```python
except_coco['잔존만기'] = except_coco.apply(lambda x: calculateQlDiffDays(x['ql_거래일자'], x['ql_상환일자']), axis=1)
except_coco
```

```python
def generate_cashflows(principal, coupon, maturity, freq=2):
    cashflows = [principal * coupon / freq] * (maturity * freq)
    cashflows[-1] += principal
    return cashflows

def generate_static_cashflows(principal, coupon, maturity):
    cashflows = [coupon] * maturity
    if len(cashflows) > 0:
        cashflows[-1] += principal
        return cashflows
    else:
        return [principal]

def calculate_present_values(cashflows, rate, freq=2):
    return sum([cf / (1 + rate / freq) ** i for i, cf in enumerate(cashflows)])

def calculate_time_weighted_present_values(cashflows, rate, freq=2):
    return sum([cf *  / (1 + rate / freq) ** i for i, cf in enumerate(cashflows)])



def calculate_present_values_df(principal, coupon, maturity, rate, freq=252):
    cashflows = generate_static_cashflows(principal*100, coupon, maturity)
    return calculate_present_values(cashflows, rate/100, freq)
```

```python
except_coco['현재가치'] = except_coco.apply(lambda x: calculate_present_values_df(x['발행가격'], x['일지급이자'], x['잔존만기'], x['종가수익률']), axis=1)
```
